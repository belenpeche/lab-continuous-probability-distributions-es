El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt

def generar_numeros_uniformes(bottom, ceiling, count):
    return np.random.uniform(bottom, ceiling, count)

# Llamar a la función con los conjuntos de parámetros especificados
numeros_1 = generar_numeros_uniformes(10, 15, 100)
numeros_2 = generar_numeros_uniformes(10, 60, 1000)

# Crear histogramas
plt.figure(figsize=(12, 6))

# Histograma para el primer conjunto de parámetros
plt.subplot(1, 2, 1)
plt.hist(numeros_1, bins=10, alpha=0.7, color='blue')
plt.title('Distribución Uniforme (10, 15, 100)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Histograma para el segundo conjunto de parámetros
plt.subplot(1, 2, 2)
plt.hist(numeros_2, bins=10, alpha=0.7, color='green')
plt.title('Distribución Uniforme (10, 60, 1000)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

1. Genera 1.000 números distribuidos normalmente con media de 10 y desviación típica de 1
1. Genera 1.000 números distribuidos normalmente con una media de 10 y una desviación típica de 50
2. Representa gráficamente las distribuciones de los datos generados.

Resultado esperado:



```python
from scipy.stats import norm 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def generar_numeros_normales_scipy(media, desviacion_estandar, count):
    return norm.rvs(loc=media, scale=desviacion_estandar, size=count)

# Generar números distribuidos normalmente
numeros_normales_1 = generar_numeros_normales_scipy(10, 1, 1000)
numeros_normales_2 = generar_numeros_normales_scipy(10, 50, 1000)

# Crear histogramas
plt.figure(figsize=(12, 6))

# Histograma para el primer conjunto de parámetros (media = 10, desviación estándar = 1)
plt.subplot(1, 2, 1)
plt.hist(numeros_normales_1, bins=30, alpha=0.7, color='blue')
plt.title('Distribución Normal (media=10, sigma=1, n=1000) - SciPy')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Histograma para el segundo conjunto de parámetros (media = 10, desviación estándar = 50)
plt.subplot(1, 2, 2)
plt.hist(numeros_normales_2, bins=30, alpha=0.7, color='green')
plt.title('Distribución Normal (media=10, sigma=50, n=1000) - SciPy')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

## Distribución Exponencial

1. Utilizando `numpy.random.exponential`, crea una función que devuelva una lista de números distribuidos exponencialmente con la media de 10. 

1. Utiliza la función para generar dos secuencias de números de tamaño 10 y 100.

1. 2. Traza las distribuciones como histogramas con el número de bins como 100.

El resultado debe ser como el siguiente:



```python
import numpy as np
import matplotlib.pyplot as plt

def generar_numeros_exponenciales(media, count):
    scale = media  # En numpy.random.exponential, el parámetro scale es la inversa de la tasa lambda (media)
    return np.random.exponential(scale, count)

# Generar números distribuidos exponencialmente
numeros_exponenciales_1 = generar_numeros_exponenciales(10, 10)
numeros_exponenciales_2 = generar_numeros_exponenciales(10, 100)

# Crear histogramas
plt.figure(figsize=(12, 6))

# Histograma para el primer conjunto de parámetros (media = 10, tamaño = 10)
plt.subplot(1, 2, 1)
plt.hist(numeros_exponenciales_1, bins=100, alpha=0.7, color='blue')
plt.title('Distribución Exponencial (media=10, n=10)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

# Histograma para el segundo conjunto de parámetros (media = 10, tamaño = 100)
plt.subplot(1, 2, 2)
plt.hist(numeros_exponenciales_2, bins=100, alpha=0.7, color='green')
plt.title('Distribución Exponencial (media=10, n=100)')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')

plt.tight_layout()
plt.show()
```

¿En qué se diferencian ambas distribuciones?

La media cambia, por lo que la distribución también cambia.
